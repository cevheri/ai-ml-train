16. Perform a `merge` using "client_list.csv" and "customer_id_list.csv". Assign the resulting dataframe as `clients`.
17. Perform a `merge` using `clients` and "purchase_log.csv" and limit the subset to only clients who made purchases. Assign the resulting dataframe as `detailed_sales`.
18. Use `groupby` to find the client who spent the most money on purchases. Determine how much he/she spent. HINT: save the intermediate dataframe from using `groupby` as `spenders` before applying slicing to determine the client who spent the most money on purchases.
19. (BONUS) Modify the answer to problem 3 slightly to determine exactly what items where purchased by the top spending client.

```python
# Perform a merge using "client_list.csv" and "customer_id_list.csv". Assign the resulting dataframe as clients.
client_csv = pd.read_csv('client_list.csv')
client_csv.info()
client_csv.columns = [x.upper() for x in client_csv.columns]
client_csv.info()
customer_csv = pd.read_csv('customer_id_list.csv')
customer_csv.info()
clients = pd.merge(left= client_csv, right= customer_csv, how='left', on=['LAST_NAME','FIRST_NAME','GENDER','AGE'])
clients
```

```python
# Perform a merge using clients and "purchase_log.csv" and limit the subset to only clients who made purchases. Assign the resulting dataframe as detailed_sales.
sales = pd.read_csv("purchase_log.csv")
detailed_sales = pd.merge(left=clients, right=sales, how='inner', on=['CUSTOMER_ID'])
detailed_sales

```

```python
# Use groupby to find the client who spent the most money on purchases. Determine how much he/she spent. HINT: save the intermediate dataframe from using groupby as spenders before applying slicing to determine the client who spent the most money on purchases.
spenders = detailed_sales.groupby(['CUSTOMER_ID','FIRST_NAME','LAST_NAME','GENDER','AGE'], as_index=False)['PRICE'].sum()
spenders[spenders.PRICE==spenders.PRICE.max()]
```

```python
# (BONUS) Modify the answer to problem 3 slightly to determine exactly what items where purchased by the top spending client.
top_spender_id = spenders[spenders.PRICE==spenders.PRICE.max()].reset_index(drop=True).loc[0,'CUSTOMER_ID']
sales.loc[sales.CUSTOMER_ID==top_spender_id, 'ITEM_DESCRIPTION']
```
