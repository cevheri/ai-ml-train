```python
import pandas as pd
```

```python
df = pd.read_csv('corona_dat.csv')
df.sample(5)
```

```python
df3 = df[['date', 'China', 'Italy', 'Brazil', 'Spain', 'US', 'Turkey']]
df3.sample(5)
```

```python
df3.set_index('date', inplace=True)
```

```python
total = df3.cumsum(axis=0)
total.sample(5)
```

```python
import bar_chart_race as bcr
```

```python
bcr.bar_chart_race(total, filename='covid19.mp4', figsize=(10,8), title='Cevheri')
```
